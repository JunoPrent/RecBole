```python
import json
import numpy as np
```

```python
with open(f"predictions_base_bpr.json") as f:
    predictions_base = json.load(f)

with open(f"predictions_cp100_l098.json") as f:
    predictions_cp = json.load(f)

with open(f"predictions_item020.json") as f:
    predictions_item = json.load(f)
```

```python
print(predictions_base)
```

```python
# Percentage of remaining items analysis
total_avg = []
for user_idx in sorted(predictions_base.keys()):
    total_avg.append(np.isin(predictions_cp[user_idx], predictions_base[user_idx]).mean())

print(np.array(total_avg).mean())

total_avg = []
for user_idx in sorted(predictions_base.keys()):
    total_avg.append(np.isin(predictions_item[user_idx], predictions_base[user_idx]).mean())

print(np.array(total_avg).mean())
```

```python
# Occurences of each item in base predictions
occurences = {i: 0 for i in range(1, 3707)}
for user_idx in sorted(predictions_base.keys()):
    for i in predictions_base[user_idx]:
        occurences[i] += 1

avg_occurences = 60400/3706
print(f"average occurences per item: {avg_occurences}")
```

```python
## Removed items analysis
items_removed = {i: 0 for i in range(1, 3707)}
for user_idx in sorted(predictions_base.keys()):
    for item_idx in predictions_base[user_idx]:
        if item_idx not in predictions_item[user_idx]:
            items_removed[item_idx] += 1

sorted_dict = dict(sorted(items_removed.items(), key=lambda item: item[1], reverse=True))
total_removed = sum(items_removed.values())
total_count, total_keys = 0, 0
for key in sorted_dict.keys():
    total_count += sorted_dict[key]
    total_keys += 1
    if total_count >= total_removed * 0.8:
        break

print(total_keys)
```

```python
## Added items analysis
items_added = {i: 0 for i in range(1, 3707)}
for user_idx in sorted(predictions_base.keys()):
    for item_idx in predictions_item[user_idx]:
        if item_idx not in predictions_base[user_idx]:
            items_added[item_idx] += 1

sorted_dict = dict(sorted(items_added.items(), key=lambda item: item[1], reverse=True))
total_added = sum(items_added.values())
total_count, top_items = 0, []
for key in sorted_dict.keys():
    total_count += sorted_dict[key]
    top_items.append(key)
    if total_count >= total_added * 0.8:
        break

print(np.array([occurences[i] for i in top_items]).mean())
print(np.array([occurences[i] + items_added[i] for i in top_items]).mean())
print(len(top_items))
```
